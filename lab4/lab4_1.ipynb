```python
import numpy as np
import os
import pandas as pd
import timeit
```

```python
def read_file(path):
    try:
        df = pd.read_csv(path, sep=";", na_values=['?'])
        df['Date'] = pd.to_datetime(df['Date'], format='%d/%m/%Y')
        df['Time'] = pd.to_datetime(df['Time'], format='%H:%M:%S').dt.time 
        df = df.drop(df.loc[(df['Sub_metering_1'] == 0) & (df['Sub_metering_2'] == 0) & (df['Sub_metering_3'] == 0)].index).dropna() 
        return df
        
    except FileNotFoundError:
        print(f"Файл '{path}' не знайдено.")
    except Exception as e:
        print(f"Помилка при відкритті файлу: {e}")
    
df = read_file('.\\household_power_consumption.txt')
```

Перетворення на масив numpy

```python
npdata = df.to_numpy()
```

1. Обрати всі домогосподарства, у яких загальна активна споживана
потужність перевищує 5 кВт.

```python
def power_filter(df):
    result = df[df['Global_active_power'] > 5][['Date', 'Time', 'Global_active_power']]
    return result

time = timeit.timeit(lambda: power_filter(df), number=1)
print('Час виконання функції: ', time)

power_filter(df)
```

```python
def power_filter_np(npdf):
    result = npdf[npdf[:, 2] > 5][:, [0, 2]]
    return result

time = timeit.timeit(lambda: power_filter_np(npdata), number=1)
print('Час виконання функції: ', time)

pd.DataFrame(power_filter_np(npdata), columns=['Date', 'Global_active_power'])
```

2. Обрати всі домогосподарства, у яких вольтаж перевищую 235 В.

```python
def voltage_filter(df):
    result = df[df['Voltage'] > 235][['Date', 'Time', 'Voltage']]
    return result

time = timeit.timeit(lambda: voltage_filter(df), number=1)
print('Час виконання функції: ', time)

voltage_filter(df)
```

```python
def voltage_filter_np(npdf):
    result = npdf[npdf[:, 4] > 235][:, [0, 4]]
    return result

time = timeit.timeit(lambda: voltage_filter_np(npdata), number=1)
print('Час виконання функції: ', time)

pd.DataFrame(voltage_filter_np(npdata), columns=['Date', 'Voltage'])
```

3. Обрати всі домогосподарства, у яких сила струму лежить в межах
19-20 А, для них виявити ті, у яких пральна машина та холодильних
споживають більше, ніж бойлер та кондиціонер.

```python
def intensity_filter(df):
    temp = df[df['Global_intensity'].between(19, 20)][['Global_intensity', 'Sub_metering_2', 'Sub_metering_3']]
    result = temp[temp['Sub_metering_2'] > temp['Sub_metering_3']]
    return result

time = timeit.timeit(lambda: intensity_filter(df), number=1)
print('Час виконання функції: ', time)

intensity_filter(df)
```

```python
def intensity_filter_np(npdf):
    temp = npdf[(npdf[:, 5] >= 19) & (npdf[:, 5] <= 20)]
    result = temp[temp[:, 7] > temp[:, 8]][:, [5, 7, 8]]
    return result

time = timeit.timeit(lambda: intensity_filter_np(npdata), number=1)
print('Час виконання функції: ', time)

pd.DataFrame(intensity_filter_np(npdata), columns=['Global_intensity', 'Sub_metering_2', 'Sub_metering_3'])
```

4. Обрати випадковим чином 500000 домогосподарств (без повторів
елементів вибірки), для них обчислити середні величини усіх 3-х
груп споживання електричної енергії.

```python
def random_rows_filter(df, num):
    temp = df.sample(num, replace=False)
    result = temp[['Sub_metering_1', 'Sub_metering_2', 'Sub_metering_3']].mean()
    return result

n = 5000

time = timeit.timeit(lambda: random_rows_filter(df, n), number=1)
print('Час виконання функції: ', time)

random_rows_filter(df, n)
```

```python
def random_rows_filter_np(npdf, num):
    ids = np.random.choice(npdf.shape[0], num, replace=False)
    temp = npdf[ids]
    result1 = np.mean(temp[:, 6])
    result2 = np.mean(temp[:, 7])
    result3 = np.mean(temp[:, 8])
    return result1, result2, result3

n=5000

time = timeit.timeit(lambda: random_rows_filter_np(npdata, n), number=1)
print('Час виконання функції: ', time)

print('Середнє зачення Sub_metering_1 %s\nСереднє зачення Sub_metering_2 %s\nСереднє зачення Sub_metering_3 %s' % random_rows_filter_np(npdata, n))
```

5. Обрати ті домогосподарства, які після 18-00 споживають понад 6
кВт за хвилину в середньому, серед відібраних визначити ті, у яких
основне споживання електроенергії у вказаний проміжок часу
припадає на пральну машину, сушарку, холодильник та освітлення
(група 2 є найбільшою), а потім обрати кожен третій результат із
першої половини та кожен четвертий результат із другої половини.

```python
def complicated_filter(df):
    temp = df[(df['Time'] > pd.to_datetime('18:00:00', format='%H:%M:%S').time()) & (df['Global_active_power'] > 6)]
    temp = temp[(temp['Sub_metering_2'] > temp['Sub_metering_3']) & (temp['Sub_metering_2'] > temp['Sub_metering_1'])][['Time', 'Global_active_power', 'Sub_metering_1', 'Sub_metering_2', 'Sub_metering_3']]
    
    half_point = len(temp) // 2
    first_half = temp.iloc[:half_point]
    second_half = temp.iloc[half_point:]

    result1 = first_half.iloc[::3]
    result2 = second_half.iloc[::4]
    result = pd.concat([result1, result2])

    return result

time = timeit.timeit(lambda: complicated_filter(df), number=1)
print('Час виконання функції: ', time)

complicated_filter(df)
```

```python
def complicated_np(npdf):
    temp = npdf[(npdf[:, 1].astype(str) > '18:00:00') & (npdf[:, 2] > 6)]
    temp = temp[(temp[:, 7] > temp[:, 6]) & (temp[:, 7] > temp[:, 8])][:, [1, 2, 6, 7, 8]]

    mid = len(temp) // 2

    result1 = temp[:mid:3]
    result2 = temp[mid::4]
    result = np.concatenate((result1, result2))
    
    return result

time = timeit.timeit(lambda: complicated_np(npdata), number=1)
print('Час виконання функції: ', time)

pd.DataFrame(complicated_np(npdata), columns=['Time', 'Global_active_power', 'Sub_metering_1', 'Sub_metering_2', 'Sub_metering_3'])
```
